## Q.1 Are movies that are more popular (operationalized as having more ratings) rated higher than movies that are less popular? [Hint: You can do a median-split of popularity to determine high vs. low popularity movies]

```python
movieNames = []
for i in range(400):
    movieNames.append(df.columns[i])
```

```python
movieNames
```

```python
nonNanCount = []
for movie in movieNames:
    nonNanCount.append(df[movie].count())
```

```python
sortednonNanCount = np.array(sorted(nonNanCount))
```

```python
medianSplitAt = np.median(sortednonNanCount)
```

```python
medianSplitAt
```

```python
dictionary = {}
dictionary['Movies'] = movieNames
dictionary['Number of Ratings'] = nonNanCount
dictionary
```

```python
popularity_df = pd.DataFrame(data = dictionary)
```

```python
popularity_df
```

```python
popularity_df['Popular'] = np.where(popularity_df['Number of Ratings'] > medianSplitAt, True, False)
```

```python
popularity_df
```

```python
medianArray = []
for i in movieNames:
    median = np.nanmedian(df[i])
    medianArray.append(median)
popularity_df['Medians'] = medianArray
```

```python
popularity_df
```

## Hypothesis Testing

### Null Hypothesis: Popular movies are not rated higher than less popular ones.

```python
results1 = stats.mannwhitneyu(popularity_df[popularity_df['Popular'] == True]['Medians'], popularity_df[popularity_df['Popular'] == False]['Medians'])
results1
```

```python
sns.set_style("whitegrid")
ax = sns.boxplot(x="Popular", y="Medians",              
data=popularity_df)
sns.set(rc={'figure.figsize':(5,5)})
```

### Q.4 What proportion of movies are rated differently by male and female viewers?

```python
male_df = df[df['Gender identity (1 = female; 2 = male; 3 = self-described)'] == 2]
```

```python
female_df = df[df['Gender identity (1 = female; 2 = male; 3 = self-described)'] == 1]
```

```python
male_df.head()
```

### Null Hypothesis: Movies are not rated differently by male and female viewers.

```python
moviesMales = {}
moviesFemales = {}
pValues = []
for i in movieNames:
    moviesMales[i] = list(male_df[i].dropna())
    moviesFemales[i] = list(female_df[i].dropna())
for i in moviesMales:
    pValue = stats.mannwhitneyu(moviesMales[i], moviesFemales[i])[1]
    pValues.append(pValue)
countSignificant = 0
for i in pValues:
    if i < 0.005:
        countSignificant += 1
print(countSignificant/len(pValues))
```

How many of these are of inconsistent quality, as experienced by viewers? [Hint: You can use the keywords in quotation marks featured in this question to identify the movies that are part of each franchise]

```python
franchiseDict = {}
franchiseNames = ['Star Wars', 'Harry Potter', 'The Matrix', 'Indiana Jones', 'Jurassic Park', 'Pirates of the Caribbean', 'Toy Story', 'Batman']
for i in franchiseNames:
    franchiseDict[i] = []
for i in movieNames:
    for j in franchiseNames:
        if j in i:
            franchiseDict[j].append(i)
franchiseDict
```

## Hypothesis Testing

### Null Hypothesis: The movies in the franchise have the same underlying distribution

```python
# Star Wars
list1 = franchiseDict['Star Wars']
# print(list1)
starWars1 = list(df['Star Wars: Episode IV - A New Hope (1977)'].dropna())
starWars2 = list(df['Star Wars: Episode II - Attack of the Clones (2002)'].dropna())
starWars3 = list(df['Star Wars: Episode V - The Empire Strikes Back (1980)'].dropna())
starWars4 = list(df['Star Wars: Episode 1 - The Phantom Menace (1999)'].dropna())
starWars5 = list(df['Star Wars: Episode VII - The Force Awakens (2015)'].dropna())
starWars6 = list(df['Star Wars: Episode VI - The Return of the Jedi (1983)'].dropna())
resultStarWars = stats.kruskal(starWars1, starWars2, starWars3, starWars4, starWars5, starWars6)
resultStarWars
```
